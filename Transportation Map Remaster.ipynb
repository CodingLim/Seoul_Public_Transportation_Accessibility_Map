## 서울시 대중교통 접근성 지도

```python
#-*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
import folium
import pandas as pd
from folium.plugins import HeatMap
from IPython.core.display import display, HTML

display(HTML("<style> .container{width:90% !important;}</style>"))
```

```python
df_sub=pd.read_csv('metro.csv')
df_sub["계산된"]=7
df_sub.head()
```

```python
df_sub2=df_sub[["X좌표(WGS)","Y좌표(WGS)","계산된"]]
df_sub2.columns=["위도","경도","계산된"]
```

```python
df_sub2.head()
df_sub2.shape
df_sub2.isnull().any()
```

```python
df_sub2=df_sub2.dropna()
```

```python
df=pd.read_csv('bus.csv')
df.head()
```

```python
df["계산된"]=df["지나가는노선수"]*0.5
coordwgt=df[["위도","경도","계산된"]]
```

```python
coordwgt.head()
```

```python
toge=pd.concat([coordwgt,df_sub2])
toge.head()
```

```python

def generateBaseMap(default_location=[37.5279, 126.99], default_zoom_start=13):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
maper=generateBaseMap()
HeatMap(data=toge,radius=15,max_val=0.8).add_to(maper) #max val 은 빨간색(최대치)를 몇으로 둘것인지. radius는 한 포인트의 크기
maper
```
